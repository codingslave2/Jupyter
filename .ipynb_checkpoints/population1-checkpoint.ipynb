## 연령별 인구 시각화

```python
import csv
import matplotlib.pyplot as plt

with open('./datas/age_2023.csv', 'r') as f:
    data = csv.reader(f)
    next(data)
    result = [] # 연령별 인구 저장할 리스트

    for row in data:
        if '대치1동' in row[0]:
            print(row[0])
            for i in row[3:]: #3번 인덱스 ~ 100번 인덱스
                result.append(int(i.replace(",", "")))
                print(int(i.replace(",", ""))) # 천단위 구분 기호 제거

print(len(row[3:]))
# print(result)
plt.rc('font', family='Malgun Gothic') # 한글 글꼴 추가
plt.title("연령별 인구 통계")
plt.style.use("ggplot") # 격자 무늬 스타일
plt.plot(result, 'r') # result - j축값

plt.xlabel('연령')
plt.ylabel('인구')
plt.show()


```
