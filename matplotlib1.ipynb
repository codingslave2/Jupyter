## 데이터 시각화
### matplotlib 모듈 사용
### plot() 함수 - 직선 또는 꺾은선 형태의 그래프

```python
import matplotlib.pyplot as plt
plt.plot([10, 20, 30, 40]) # 리스트는 y축값이며, x값은 0부터 1씩 증가함
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [12, 43, 25, 15]
# plt.title("plotting")
plt.rc('font', family="Malgun Gothic")
plt.title("꺾은선 그래프")
plt.plot(x, y) # 앞 리스트 - x축, 뒤 리스트 - y 축
plt.show()
```

```python
plt.title("범례 사용")
plt.plot([10, 20, 30, 40], label='증가', color = 'blue')
plt.plot([40, 30, 20, 10], label='감소', color = 'red')
plt.legend()
plt.show()
```

## bar() - 막대 그래프
### 막대의 길이는 각 데이터의 크기를 의미함

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [12, 43, 25, 15]
plt.bar(x, y)
plt.show()
```

```python
plt.rc('font', family='Malgun Gothic')
plt.title("막대 그래프")

plt.bar(range(6), (1, 2, 3, 5, 6, 8))
plt.show()
```
