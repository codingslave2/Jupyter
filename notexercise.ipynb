# 서울시 운동을 하지 않는 이유 통계
## excel 파일을 읽기 -pandas.read_excel(파일)

```python
import pandas as pd
import warnings

warnings.simplefilter('ignore')

# 제목행의 1번 인덱스만 출력(본문은 숨김)
not_exercise = pd.read_excel("./datas/not_exercise.xlsx", header=1)
not_exercise
```

```python
not_exercise.columns[0]
```

```python
# 칼럼 삭제 - 객체 drop()
not_exercise.drop(columns='시점', inplace=True)
not_exercise
```

```python
# 칼럼명 변경 - 객체, rename()
not_exercise.rename(columns={'구분별(1)':'대분류', '구분별(2)':'분류'}, inplace=True)
not_exercise
```

```python
# 행 삭제 - 객체.drop(index=range(시작행, 끝행))
not_exercise.drop(index=range(22, 52), inplace=True)
not_exercise
```

```python
# 2행의 NaN을 '성별'로 변경
not_exercise.loc[2, '대분류'] = '성별'
not_exercise
```

```python
not_ex_sex = not_exercise[not_exercise['대분류'] == '성별'].copy() # 깊은 복사
not_ex_sex
```

```python
# 대분류 칼럼 삭제
not_ex_sex.drop(columns='대분류', inplace=True)
not_ex_sex
```

```python
# index를 '분류'로 세팅
not_ex_sex.set_index('분류', inplace=True)
not_ex_sex
```

## figure : 그래프를 그릴 공간(종이)
## ax(axis) : 그 공간 중 내가 가용할 부분

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2)

plt.rc('font', family='Malgun Gothic')
not_ex_sex['운동을 할 충분한 시간이 없어서'].plot.pie(ax=ax[0], autopct='%.1f%%', startangle=90)
ax[0].set_title('운동을 할 충분한 시간이 없어서')
ax[0].set_ylabel('')

not_ex_sex['운동을 싫어해서'].plot.pie(ax=ax[1], autopct='%.1f%%', startangle=90)
ax[1].set_title('운동을 싫어해서')
ax[1].set_ylabel('')

plt.show()
```
